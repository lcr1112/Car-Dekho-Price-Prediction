# 2. Feature Engineering and Feature Selection
## converting categorical varibles into numerical

```python
Seller_type=pd.get_dummies(df['Seller_Type'])
Transmission=pd.get_dummies(df['Transmission'])
Fuel_type=pd.get_dummies(df['Fuel_Type'])
```

```python
Seller_type
```

```python
Transmission
```

```python
Fuel_type
```

```python
df
```

```python
df.columns
```

```python
df.drop(['Car_Name', 'Year','Fuel_Type','Seller_Type','Transmission'],axis=1,inplace=True)
```

```python
df
```

```python
df1=pd.concat([df,Seller_type,Transmission,Fuel_type],axis=1)
```

```python
df1
```

```python
y=df1['Selling_Price']
```

```python
y
```

```python
x=df1.drop(['Selling_Price'],axis=1)
```

```python
x
```

# 3. Train Test Split, Creating &Training Model and Model Evaluation
## Using Linear Regression

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
model.score(x_test,y_test)
```

```python
from sklearn import metrics
```

```python
MAE=metrics.mean_absolute_error(y_test,y_predict)
MSE=metrics.mean_squared_error(y_test,y_predict)
R2_score=metrics.r2_score(y_test,y_predict)
```

```python
print(f'MAE:{MAE}\n MSE:{MSE}\n R2_score:{R2_score}')
```

```python
sns.distplot(y_test,color='b',label='y_test')
sns.distplot(y_predict,color='r',label='y_pred.')
plt.legend()
```
